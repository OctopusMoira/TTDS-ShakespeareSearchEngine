```python
import unicodedata
def strip_accents(s):
   return ''.join(c for c in unicodedata.normalize('NFD', s)
                  if unicodedata.category(c) != 'Mn')
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
uniqueId2stemmedList = dict()
for uniqueId in uniqueId2stripped_content.keys():
    stemmedList = []
    content = uniqueId2stripped_content[uniqueId]
    for word in content.strip().split(" "):
        stemmedList.append(stemmer.stem(word))
    uniqueId2stemmedList[uniqueId] = stemmedList
```

```python
stem2id_count = dict()
```

```python
for uniqueId in uniqueId2stemmedList.keys():
    thisList = uniqueId2stemmedList[uniqueId]
    for keyStem in thisList:
        if keyStem not in stem2id_count.keys():
            newid2c = dict()
            newid2c[uniqueId] = 1
            stem2id_count[keyStem] = newid2c
            continue
        if uniqueId not in stem2id_count[keyStem].keys():
            stem2id_count[keyStem][uniqueId] = 1
            continue
        stem2id_count[keyStem][uniqueId] += 1
```

```python
def FreqDocOfTerm(term):
    f = 0
    if term in stem2id_count.keys():
        f = len(stem2id_count[term])
    return f
```

```python
DocCountN = len(uniqueId2stemmedList.keys())
```

```python
import math
def IDF(term):
    fterm = FreqDocOfTerm(term)
    return math.log10(1+(DocCountN - fterm + 0.5)/(fterm+0.5))
```

```python
def FreInDOfTerm(term, dID):
    if term in stem2id_count.keys():
        if dID in stem2id_count[term].keys():
            return stem2id_count[term][dID]
    return 0
```

```python
avgFieldLen = 0
for uid in uniqueId2stemmedList.keys():
    avgFieldLen += len(uniqueId2stemmedList[uid])
avgFieldLen /= DocCountN
```

```python
avgFieldLen
```

```python
def processQuery(query):
#     resultId2countsum = dict()
    cleanedQ = []
    query = query.translate(str.maketrans(string.punctuation, replacement))
    for word in query.strip().split(" "):
        foldW = word.casefold()
        if foldW not in stopw:
            stemW = stemmer.stem(foldW)
            if stemW in stem2id_count.keys():
                cleanedQ.append(stemW)
    return cleanedQ
#                 for uid in stem2id_count[stemW].keys():
#                     thisC = stem2id_count[stemW][uid]
#                     if uid not in resultId2countsum.keys():
#                         resultId2countsum[uid] = thisC
#                         continue
#                     resultId2countsum[uid] += thisC
#     return resultId2countsum
# processQuery(querytest)
```

```python
def BM25(docID, query):
    bm25summ = 0
    for term in query:
        fieldLen = len(query)
        if term in stem2id_count.keys():
            finDt = FreInDOfTerm(term, docID)
            bm25summ += ( IDF(term) * (finDt * 2.2) / ( finDt + 1.2 * (0.25 + 0.75 * ( fieldLen / avgFieldLen)) ) )
    return bm25summ
```

```python
querytest = "love hate money power"
```

```python
cleanQ = processQuery(querytest)
uid2bm25 = dict()
for uid in uniqueId2stemmedList.keys():
    uid2bm25[uid] = BM25(uid, cleanQ)
```
