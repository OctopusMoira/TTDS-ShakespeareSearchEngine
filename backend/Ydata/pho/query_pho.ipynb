```python
pho2term = np.load('./pho2term.npy',allow_pickle='TRUE').item()
```

```python
insertlist = []
for key in pho2term.keys():
    if key != '':
        tokenbox = pho2term[key].split(" ")
        newinsert = dict()
        newinsert['sound'] = key
        newinsert['tokenbox'] = tokenbox
        insertlist.append(newinsert)
```

```python
phoboxC = searchindb['phobox']
phoboxC.insert_many(insertlist)
```

```python
def phoquery(word):
    word_pho = phonetics.metaphone(word)
    thispho = phoboxC.find_one({'sound': word_pho})
    if thispho:
        return thispho['tokenbox']
    return []
```

```python
phoquery('killin')
```

```python
def new_q(query):
    query_pho = []
    for word in query.split():
        word_pho = phonetics.metaphone(word)
        query_pho.append(word_pho)
    new_q = []
    for term in query_pho:
        for w in pho2term.keys():
            if w == term:
                new_q += pho2term[w].split() 
    return list(set(new_q))
```
